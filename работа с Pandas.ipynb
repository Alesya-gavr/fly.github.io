```python
#рассчитываем разницу между средней реакцией игроков максимального возраста и минимального возраста
react_old=players[players.Age.max()==players.Age].Reactions.mean()
react_young=players[players.Age.min()==players.Age].Reactions.mean()
diff=react_old-react_young
diff
```

```python
#во сколько раз сила выстрела у игроков с максимальной агрессией выше, чем у игроков с минимальной агрессией
agr_max=players[players.Aggression==players.Aggression.max()].ShotPower.mean()
agr_min=players[players.Aggression==players.Aggression.min()].ShotPower.mean()
agr_max/agr_min
```

```python
#выводим минимальный возраст игроков у которых реакция и хладнокровие составляют 90% от максимальных показателей
players[(players.Composure>players.Composure.max()*0.9)&(players.Reactions>players.Reactions.max()*0.9)].Age.min()
```
